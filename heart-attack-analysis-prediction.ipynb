## Data Content

* Age: Hastanın yaşı
* Sex: Hastanın cinsiyeti
* xang: egzersize bağlı göğüs ağrısı (1 = yes; 0 = no)
* ca: büyük damar sayısı (0-3)
* cp: Göğüs ağrısı tipi
    * Value 1: tipik göğüs ağrısı
    * Value 2: tipik olmayan göğüs ağrısı
    * Value 3: non-anginal pain
    * Value 4: asymptomatic
* trtbps: dinlenme kan basıncı (in mm Hg)
* chol: kolestrol in mg/dl fetched via BMI sensor
* fbs: (açken kan şekeri > 120 mg/dl) (1 = true; 0 = false)
* rest_ecg: resting electrocardiographic results
    * Value 0: normal
    * Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    * Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
* thalach: maksimum kalp atış hızı
* target: 0= daha az kalp krizi şansı 1= daha fazla kalp krizi şansı

## Outlier Detection

* Aykırı değer yakalama 
* 


```python
numeric_list = ["age", "trtbps","chol","thalachh","oldpeak"]
df_numeric = df.loc[:, numeric_list]
df_numeric.head()
```

```python
df.describe()
```

```python
# aykırı değerleri bulma ve çıkarma
for i in numeric_list:
    
    # IQR
    Q1 = np.percentile(df.loc[:, i],25)
    Q3 = np.percentile(df.loc[:, i],75)
    
    IQR = Q3 - Q1
    
    print("Old shape: ", df.loc[:, i].shape)
    
    # upper bound
    upper = np.where(df.loc[:, i] >= (Q3 +2.5*IQR))
    
    # lower bound
    lower = np.where(df.loc[:, i] <= (Q1 - 2.5*IQR))
    
    print("{} -- {}".format(upper, lower))
    
    try:
        df.drop(upper[0], inplace = True)
    except: print("KeyError: {} not found in axis".format(upper[0]))
    
    try:
        df.drop(lower[0], inplace = True)
    except:  print("KeyError: {} not found in axis".format(lower[0]))
        
    print("New shape: ", df.shape)
```

## Modelling

```python
df1 = df.copy()
```

```python
df1 = pd.get_dummies(df1, columns = categorical_list[:-1], drop_first = True)
df1.head()
```

```python
X = df1.drop(["output"], axis = 1)
y = df1[["output"]]
```

### Scaling

```python
scaler = StandardScaler()
scaler
```

```python
#numeric değerleri standardizasyon işlemi uyguladık
X[numeric_list[:-1]] = scaler.fit_transform(X[numeric_list[:-1]])
X.head()
```

### Train/Test Split

```python
#%90 a %10 olmak üzere eğitim ve test olarak ayırdık
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 3)
print("X_train: {}".format(X_train.shape))
print("X_test: {}".format(X_test.shape))
print("y_train: {}".format(y_train.shape))
print("y_test: {}".format(y_test.shape))
```

### Logistic Regression

```python
logreg = LogisticRegression()
logreg
```

```python
# fitting = training(algoritmayı eğitiyoruz)
logreg.fit(X_train, y_train)
```

```python
# calculate probabilities( olasılık olarak hesaplanan değerler)
y_pred_prob = logreg.predict_proba(X_test)
y_pred_prob
```

```python
# olasılık olan değerleri 1-0 olarak dönüştürme
y_pred = np.argmax(y_pred_prob, axis = 1)
y_pred
```

```python
#dummy_ = pd.DataFrame(y_pred_prob)
#dummy_["y_pred"] = y_pred
#dummy_.head()
```

```python
# algoritmanın çalışma doğruluğu %90
print("Test accuracy: {}".format(accuracy_score(y_pred, y_test)))
```

```python
# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob[:,1])
```

```python
# plot curve
plt.plot([0,1],[0,1],"k--")
plt.plot(fpr, tpr, label = "Logistic Regression")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Logistic Regression ROC Curve")
plt.show()
```

### Logistic Regression Hyperparameter Tuning

```python
lr = LogisticRegression()
lr
```

```python
penalty = ["l1", "l2"]

parameters = {"penalty":penalty}
```

```python
lr_searcher = GridSearchCV(lr, parameters)
```

```python
lr_searcher.fit(X_train, y_train)
```

```python
print("Best parameters: ",lr_searcher.best_params_)
```

```python
y_pred = lr_searcher.predict(X_test)
```

```python
print("Test accuracy: {}".format(accuracy_score(y_pred, y_test)))
```
